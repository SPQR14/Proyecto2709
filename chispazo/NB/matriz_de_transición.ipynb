```python
import pandas as pd
#transiciones = ['A', 'B', 'B', 'C', 'B', 'A', 'D', 'D', 'A', 'B', 'A', 'D']
transiciones = [1, 2,2,3,2,1,4,4,1,2,1,4, 1, 1, 1, 1, 2, 3, 4, 4, 4, 5, 2, 2, 2, 3, 1, 4, 2, 3, 5, 1]
df = pd.DataFrame(transiciones)
```

```python
df
```

```python
df['shift'] = df[0].shift(-1)
```

```python
df['count'] = 1
```

```python
mat_tran = df.groupby([0, 'shift']).count().unstack().fillna(0)
```

```python
mat_tran
```

```python
mat_tran = mat_tran.div(mat_tran.sum(axis = 1), axis = 0).values
```

```python
df = pd.DataFrame(mat_tran)
```
