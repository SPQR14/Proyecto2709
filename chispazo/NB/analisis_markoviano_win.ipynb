```python
import pandas as pd
import numpy as np
from random import seed
from random import random
import matplotlib as plt
import platform
```

```python
if(platform.system() == 'Windows'):
    df = pd.read_csv("..\\Data_set\\real\\Chispazo.csv", sep = ',', encoding = 'utf8')
else:
    df = pd.read_csv("../Data_set/real/Chispazo.csv", sep = ',', encoding = 'utf8')
```

```python
df.head()
```

```python
df.sort_values('CONCURSO', ascending = True, inplace = True)
```

```python
df.reset_index(inplace = True)
```

```python
df = df[['FECHA','CONCURSO', 'R1', 'R2', 'R3', 'R4', 'R5']]
```

```python
df.head()
```

```python
matriz_R1 = list(df['R1'])
```

```python
matriz_R1 = pd.DataFrame(matriz_R1)
matriz_R1.head()
```

```python
matriz_R1['shift'] = matriz_R1[0].shift(-1)
```

```python
matriz_R1['count'] = 1
```

```python
matriz_de_transicion = matriz_R1.groupby([0, 'shift']).count().unstack().fillna(0)
```

```python
matriz_de_transicion = matriz_de_transicion.div(matriz_de_transicion.sum(axis = 1), axis = 0).values
```

```python
matriz_de_transicion = pd.DataFrame(matriz_de_transicion)
```

```python
matriz_de_transicion
```

```python
matriz_de_transicion = matriz_de_transicion.rename(columns = dict(zip(matriz_de_transicion.columns, [x for x in range(min(list(df['R1'])), max(list(df['R1'])) + 1)])))
```

```python
matriz_de_transicion.insert(0, 'estado', [x for x in range(min(list(df['R1'])), max(list(df['R1'])) + 1)])
```

```python
matriz_de_transicion
```

```python
matriz_de_transicion.to_csv('R1.csv', sep = ',', encoding = 'utf8')
```
